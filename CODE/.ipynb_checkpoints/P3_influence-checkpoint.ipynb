```python
import networkx as nx
import pandas as pd
```

```python
data = pd.read_csv('influence_data.csv')
```

```python
data
```

```python
genres=set(data['influencer_main_genre'])
edge=pd.DataFrame(index=genres,columns=genres,dtype=float).fillna(value = 0)
size=pd.DataFrame(index=genres,columns=['size'],dtype=float).fillna(value = 0)
```

```python
G=nx.DiGraph()
for index, row in data.iterrows():
    if row['influencer_main_genre']!=row['follower_main_genre']:
        edge.loc[row.influencer_main_genre,row.follower_main_genre]=edge.loc[row.influencer_main_genre,row.follower_main_genre]+1
    else:
        size.loc[row.influencer_main_genre,'size']=size.loc[row.influencer_main_genre,'size']+1
```

```python
size
```

```python
size.to_csv('size.csv')
edge.to_csv('edge.csv')
```
