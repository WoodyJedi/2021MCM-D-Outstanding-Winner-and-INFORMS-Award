```python
import networkx as nx
import pandas as pd
import numpy as np
```

```python
influence_data = pd.read_csv('influence_data.csv')
influence_data
```

```python
Artist_embeddings = pd.read_csv('Artist_embeddings.csv',index_col=0)
Artist_embeddings
```

```python
data_name=['Component_0','Component_1','Component_2']
```

```python
influencer_0 = []
influencer_1 = []
influencer_2 = []
influencer_data=[influencer_0,influencer_1,influencer_2]
```

```python
follower_0 = []
follower_1 = []
follower_2 = []
follower_data=[follower_0,follower_1,follower_2]
```

```python
for index, row in list(influence_data.iterrows()):
    for i in range(3):
        try:
            Artist_embeddings.loc[row.influencer_id,data_name[i]]
            Artist_embeddings.loc[row.follower_id,data_name[i]]
        except:
            continue
        influencer_data[i].append(Artist_embeddings.loc[row.influencer_id,data_name[i]])
        follower_data[i].append(Artist_embeddings.loc[row.follower_id,data_name[i]])
```

```python
result_correlation = pd.DataFrame(index=data_name,columns=['correlation']).fillna(value=0)
INDEX=range(len(follower_data[0]))
for i in range(len(data_name)):
    dinfluencer = pd.Series(influencer_data[i], index=INDEX,dtype=float)
    dfollower = pd.Series(follower_data[i], index=INDEX,dtype=float)
    df=pd.DataFrame({'influencer':dinfluencer,'follower':dfollower},dtype=float)
    result_correlation.iloc[i,0]=df.corr(method='pearson').iloc[0,1]
print(result_correlation)
```

```python
result_correlation.to_csv('cooooo_pearson_correlation_of_connected_musicians.csv')
```
